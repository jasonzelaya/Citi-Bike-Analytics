```python
# Import Dependencies
import os
import pandas as pd
```

```python
# View raw data files to aggregate
os.listdir('data/raw_data')
```

```python
# Read Citi Bike data into Pandas data frames
january = pd.read_csv('data/raw_data/201901-citibike-tripdata.csv')
february = pd.read_csv('data/raw_data/201902-citibike-tripdata.csv')
march = pd.read_csv('data/raw_data/201903-citibike-tripdata.csv')
april = pd.read_csv('data/raw_data/201904-citibike-tripdata.csv')
may_2019 = pd.read_csv('data/raw_data/201905-citibike-tripdata.csv')
june = pd.read_csv('data/raw_data/201906-citibike-tripdata.csv')
july = pd.read_csv('data/raw_data/201907-citibike-tripdata.csv')
august = pd.read_csv('data/raw_data/201908-citibike-tripdata.csv')
september = pd.read_csv('data/raw_data/201909-citibike-tripdata.csv')
october = pd.read_csv('data/raw_data/201910-citibike-tripdata.csv')
november = pd.read_csv('data/raw_data/201911-citibike-tripdata.csv')
december = pd.read_csv('data/raw_data/201912-citibike-tripdata.csv')
```

```python
# Store read Citi Bike data into separate data frames to avoid rereading the data
jan = january
feb = february
mar = march
apr = april
may = may_2019
jun = june
jul = july
aug = august 
sep = september
octo = october
nov = november
dec = december
```
