```python
# Import Dependencies
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# View raw data files to aggregate
os.listdir('data/raw_data')
```

```python
# Tableau Public can work with a maximum of only 15,000,000 rows. The total rows for the full year of 
# 2019 is 20,523,930 so only the summer and winter months data will be used
# Read Citi Bike data into Pandas data frames
jan = pd.read_csv('data/raw_data/201901-citibike-tripdata.csv')
feb = pd.read_csv('data/raw_data/201902-citibike-tripdata.csv')
jun = pd.read_csv('data/raw_data/201906-citibike-tripdata.csv')
jul = pd.read_csv('data/raw_data/201907-citibike-tripdata.csv')
aug = pd.read_csv('data/raw_data/201908-citibike-tripdata.csv')
dec = pd.read_csv('data/raw_data/201912-citibike-tripdata.csv')
```

```python
# Preview data frames
display(jan.head())
display(feb.head())
display(jun.head())
display(jul.head())
display(aug.head())
display(dec.head())
```

```python
# Separate words with "_" in column names to make them easier to read
col_names = ['trip_duration', 'start_time', 'stop_time', 'start_station_id', 'start_station_name', \
                 'start_station_latitude', 'start_station_longitude', 'end_station_id', \
                 'end_station_name', 'end_station_latitude', 'end_station_longitude', 'bike_id', \
                 'user_type', 'birth_year', 'gender']

jan.columns = col_names
feb.columns = col_names
jun.columns = col_names
jul.columns = col_names
aug.columns = col_names
dec.columns = col_names
```

```python
# Check how many "-" values are in each data frame
print(jan[jan.isin(["-"])].count().sum())
print(feb[feb.isin(["-"])].count().sum())
print(jun[jun.isin(["-"])].count().sum())
print(jul[jul.isin(["-"])].count().sum())
print(aug[aug.isin(["-"])].count().sum())
print(dec[dec.isin(["-"])].count().sum())
```

```python
# Replace infinity values with NaN
jan.replace([np.inf, -np.inf], np.nan, inplace=True)
feb.replace([np.inf, -np.inf], np.nan, inplace=True)
jun.replace([np.inf, -np.inf], np.nan, inplace=True)
jul.replace([np.inf, -np.inf], np.nan, inplace=True)
aug.replace([np.inf, -np.inf], np.nan, inplace=True)
dec.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Drop NaNs
jan.dropna(inplace=True)
feb.dropna(inplace=True)
jun.dropna(inplace=True)
jul.dropna(inplace=True)
aug.dropna(inplace=True)
dec.dropna(inplace=True)
```

```python
# Check if data types for each corresponding column are the same among all data frames
# Concatenate all of the data frames to compare data types
data_types_df = pd.concat([jan.dtypes, feb.dtypes, jun.dtypes, jul.dtypes, aug.dtypes, dec.dtypes], \
                              axis=1)

# Rename columns to make the data frame easier to understand
data_types_df.columns = ['jan', 'feb', 'jun', 'jul', 'aug', 'dec']

# Check for inconsistent data types in corresponding columns
data_types_df
```

```python
# Data types consistency check conclusion:
# start_station_id and end_station_id columns' data types are inconsistent
# start_station_id - 4 data frames = float64, 7 data frames = int64
# end_station_id - 4 data frames = float64, 7 data frames = int64

# Solution: Convert the 4 float64 data frames to int64 since all start/end station IDs are whole numbers
```

```python
# Convert the columns data types for consistency across all data frames
jan_converted = jan.astype({'start_station_id': np.int64, 'end_station_id': np.int64})
feb_converted = feb.astype({'start_station_id': np.int64, 'end_station_id': np.int64})
jul_converted = jul.astype({'start_station_id': np.int64, 'end_station_id': np.int64})
aug_converted = aug.astype({'start_station_id': np.int64, 'end_station_id': np.int64})
```

```python
# Confirm matching data types
# Concatenate the converted data frames with the data frames that did not need to be converted to compare
# new data types
data_types_converted = pd.concat([jan_converted.dtypes, feb_converted.dtypes, jun.dtypes, \
                                  jul_converted.dtypes, aug_converted.dtypes, dec.dtypes], axis=1)

# Rename columns to make data frame easier to understand
data_types_converted.columns = ['jan', 'feb', 'jun', 'jul', 'aug', 'dec']

# Confirm matching data types
data_types_converted
```

```python
# Confirm 'user_type' column has only two unique values: "Customer" and "Subscriber"
print(jan_converted['user_type'].unique())
print(feb_converted['user_type'].unique())
print(jun['user_type'].unique())
print(jul_converted['user_type'].unique())
print(aug_converted['user_type'].unique())
print(dec['user_type'].unique())
```

```python
# Confirm 'gender' column in each data frame has only three unique values: "0", "1", and "2"
print(jan_converted['gender'].unique())
print(feb_converted['gender'].unique())
print(jun['gender'].unique())
print(jul_converted['gender'].unique())
print(aug_converted['gender'].unique())
print(dec['gender'].unique())
```

```python
# Aggregate all data
all_data = pd.concat([jan_converted, feb_converted, jun, jul_converted, aug_converted, dec])
```

```python
# Preview aggregated data frame
all_data.head()
```

```python
# Confirm the ages in the data make sense
print(f'Youngest rider\'s age and birth year in the data: \
        {max(all_data["birth_year"])}, {2019 - max(all_data["birth_year"])}')
print(f'Oldest rider\'s age and birth year in the data: \
        {min(all_data["birth_year"])}, {2019 - min(all_data["birth_year"])}')



# Use the below code if you want the age confirmations to update every year, but the Tableau data story is based on the 
#     year 2019
# print(f'Youngest rider\'s age and birth year in the data: \
#         {max(all_data["birth_year"])}, {datetime.now().year - max(all_data["birth_year"])}')
# print(f'Oldest rider\'s age and birth year in the data: \
#         {min(all_data["birth_year"])}, {datetime.now().year - min(all_data["birth_year"])}')
```

```python
# Oldest age, 134 years (as of 2019), does not make sense so 'birth_year' column will be filtered
# Google search returns several articles discussing increase in U.S. bike trips by seniors through age 79
# so 79 years will be maximum age for Citi Bike data
all_data = all_data[all_data['birth_year'] >= 1940]
```

```python
# Confirm total rows <= 15,000,000 to make sure Tableau Public will be able to work with this data
all_data['birth_year'].count()
```

```python
# Export aggregated data to csv file
all_data.to_csv('data/citi_bike_summer_winter_2019.csv', sep=',', index=False)
```
